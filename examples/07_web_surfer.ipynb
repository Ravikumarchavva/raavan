# 07 — WebSurfer: Agentic Web Browsing

Demonstrates `WebSurferTool` which drives a **Playwright** browser:
- Navigate to URLs
- Extract text / markdown
- Take full-page screenshots
- Execute JavaScript
- Scroll, click, go back

**Prerequisites**: `playwright` installed and browsers downloaded  
(`uv run playwright install chromium`).

```python
from raavan.extensions.tools.web_surfer import WebSurferTool
```

## Basic browsing actions

Set `headless=False` to watch the browser window open.

```python

async def demo():
    surfer = WebSurferTool(headless=True, browser_type='chromium')
    try:
        # Navigate
        result = await surfer.execute(action='navigate', url='https://github.com/Ravikumarchavva/raavan')
        print('Navigate:', result.content[0]['text'])

        # Metadata
        result = await surfer.execute(action='get_metadata')
        print('Metadata:', result.content[0]['text'][:200])

        # Markdown extract
        result = await surfer.execute(action='extract_markdown')
        print('Markdown:', result.content[0]['text'][:300])

        # Screenshot
        result = await surfer.execute(action='screenshot', full_page=True)
        if len(result.content) > 1 and result.content[1]['type'] == 'image':
            print(f'Screenshot: {len(result.content[1]["data"])} bytes (base64)')

        # JS execution
        result = await surfer.execute(action='execute_js', javascript="document.title")
        print('Page title (JS):', result.content[0]['text'])

    except Exception as e:
        print(f'⚠ WebSurfer unavailable: {e}')
        print('Requires: playwright install chromium  (run once after pip install playwright)')
    finally:
        try:
            await surfer.execute(action='close')
            print('Browser closed')
        except Exception:
            pass

try:
    await demo()
except Exception as e:
    print(f'⚠ Demo failed: {e}')
```
